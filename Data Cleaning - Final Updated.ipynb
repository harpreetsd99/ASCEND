```python
# Kautilya - Survey Data

def cleaned_survey_report_data(base_dir, file_names):

    # Initialize an empty list to store DataFrames
    dfs = []
    
    # Iterate through each file name
    for file_name in file_names:
        # Construct the full file path
        file_path = os.path.join(base_dir, file_name)
        
        # Check if the file exists
        if os.path.exists(file_path):
            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path, encoding='utf-8')

            # Function to adjust columns for French data
            def adjust_columns(df):
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Gender'] = df['Genre']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Age'] = df[' Âge']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Status in Canada'] = df['Statut au Canada']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Marital Status'] = df['Situation maritale']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Country of Birth'] = df['Pays de naissance']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Country you have_will immigrate from'] = df['Pays d_où vous avez immigré_immigrerez']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Year landed in Canada _enter 0 if not yet landed_'] = df['Année d_arrivée au Canada _saisir 0 si non applicable_']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Highest level of education'] = df['Plus haut niveau de formation']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'What is your current employment status'] = df['Quelle est votre situation professionnelle actuelle']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'What industry or sector are you working in'] = df['Dans quel secteur dactivite ou industrie travaillez_vous']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'Is this your preferred industry or sector'] = df['Est_ce votre secteur ou industrie de prédilection']
                df.loc[(df["Program Version"] == "French") | (df["Program Version"] == "SDECB"), 'What is your current annual employment salary'] = df['Quel est votre salaire annuel actuel']
                return df

            # Function to convert French text to English
            def french_convert(df):
                df['Age'].replace('Préfère ne pas répondre', 'Prefer not to disclose', inplace=True)
                
                df['ChapterName'].replace('Évaluation finale', 'Final Evaluation', inplace=True)
                df['ChapterName'].replace('Questionnaire d’avant-programme', 'Questionnaire', inplace=True)
                
                df['LessonName'].replace('Partie 1 – Satisfaction vis-à-vis du programme ASCEND', 'Post-program survey: Part 1 - Program satisfaction', inplace=True)
                df['LessonName'].replace('Partie 2 – Auto-évaluation de votre préparation à l’emploi', 'Post-program survey: Part 2 - Soft skills self-assessment', inplace=True)
                df['LessonName'].replace('Questionnaire d’avant-programme : Auto-évaluation de votre préparation à l’emploi', 'Pre-program survey: Soft skills self-assessment', inplace=True)
                
                df['Gender'].replace('Homme', 'Male', inplace=True)
                df['Gender'].replace('Femme', 'Female', inplace=True)
                df['Gender'].replace('Préfère ne pas répondre', 'Prefer not to disclose', inplace=True)
                
                df['Marital Status'].replace('Séparé.e / Divorcé.e', 'Separated/divorced', inplace=True)
                df['Marital Status'].replace('Célibataire', 'Single', inplace=True)
                df['Marital Status'].replace('Marié.e / En concubinage', 'Married/live with partner', inplace=True)
                
                df['Is this your preferred industry or sector'].replace('Non', 'No', inplace=True)
                df['Is this your preferred industry or sector'].replace('Oui', 'Yes', inplace=True)
            
                df['What is your current annual employment salary'].replace(r'.*25,000.*50,000.*', '$25,000 - $50,000', regex=True, inplace=True)
                df['What is your current annual employment salary'].replace("$25,000 \u2013 $50,000", "$25,000 - $50,000", inplace=True)
                df['What is your current annual employment salary'].replace('$25,000 – $50,000', '$25,000 - $50,000', inplace=True)
                df['What is your current annual employment salary'].replace('$25,000  $50,000', '$25,000 - $50,000', inplace=True)
                df['What is your current annual employment salary'].replace('Je suis actuellement sans emploi', 'I am not currently employed', inplace=True)
                df['What is your current annual employment salary'].replace('Moins de $25,000', 'Less than $25,000', inplace=True)
                df['What is your current annual employment salary'].replace('Plus de $100,000', 'More than $100,000', inplace=True)
                
                df['What industry or sector are you working in'].replace('Je suis actuellement sans emploi', 'I am not currently employed', inplace=True)
                df['What industry or sector are you working in'].replace('Commerce de détail', 'Retail trade', inplace=True)
                df['What industry or sector are you working in'].replace('Transport et entreposage', 'Transportation and warehousing', inplace=True)
                df['What industry or sector are you working in'].replace('Finance et assurances', 'Finance and insurance', inplace=True)
                df['What industry or sector are you working in'].replace('Services professionnels, scientifiques et techniques', 'Professional, scientific and technical services', inplace=True)
                df['What industry or sector are you working in'].replace('Gestion de sociétés et d’entreprises', 'Management of companies and enterprises', inplace=True)
                df['What industry or sector are you working in'].replace('Soins de santé et assistance sociale', 'Health care and social assistance', inplace=True)
                df['What industry or sector are you working in'].replace('Arts, spectacles et loisirs', 'Arts, entertainment and recreation', inplace=True)
                df['What industry or sector are you working in'].replace('Services d’hébergement et de restauration', 'Health care and social assistance', inplace=True)
                df['What industry or sector are you working in'].replace('Administration publique', 'Public administration', inplace=True)
                df['What industry or sector are you working in'].replace('Autres services', 'Other services', inplace=True)
                df['What industry or sector are you working in'].replace('Services administratifs, de support, de gestion des déchets et d’assainissement', 'Administrative and support, waste management and remediation services', inplace=True)
                
                df['What is your current employment status'].replace('A l’école ou en formation', 'In school or training', inplace=True)
                df['What is your current employment status'].replace('Autre', 'Other', inplace=True)
                df['What is your current employment status'].replace('Employé.e', 'Employed', inplace=True)
                df['What is your current employment status'].replace('En arrêt de travail (congé, maternité, paternité, etc.)', 'On leave from employment (Leave of absence, Maternity, Paternity, etc.)', inplace=True)
                df['What is your current employment status'].replace('Sans emploi', 'Unemployed', inplace=True)
                df['What is your current employment status'].replace('Travailleur.euse indépendant.e (Contractuel.le ou Consultant.e)', 'Self-employed (Contractor or Consultant)', inplace=True)
                
                df['Highest level of education'].replace('Études secondaires non terminées', 'Less than high school', inplace=True)
                df['Highest level of education'].replace('Baccalauréat', 'High school diploma or equivalent', inplace=True)
                df['Highest level of education'].replace('Diplôme', 'Diploma', inplace=True)
                df['Highest level of education'].replace('Diplôme d’études secondaires ou équivalent', 'High school diploma or equivalent', inplace=True)
                df['Highest level of education'].replace('Doctorat', 'PhD', inplace=True)
                df['Highest level of education'].replace('Formation post-secondaire non-diplômante (école ou université)', 'Some post-secondary (college or university courses)', inplace=True)
                df['Highest level of education'].replace('Maîtrise', 'Master degree', inplace=True)
                
                df['Status in Canada'].replace('Citoyen.ne canadien.nne', 'Canadian citizen', inplace=True)
                df['Status in Canada'].replace('Étudiant.e avec un permis de travail', 'Student with work visa', inplace=True)
                df['Status in Canada'].replace('Époux.se de Candidat.e principal.e à la Résidence permanente (au Canada)', 'Spouse of Permanent Resident Principal Applicant (In Canada)', inplace=True)
                df['Status in Canada'].replace('Réfugié.e au sens de la Convention', 'Convention Refugee', inplace=True)
                df['Status in Canada'].replace('Résident.e permanent.e (au Canada)', 'Permanent Resident (In Canada)', inplace=True)
                df['Status in Canada'].replace('Résident.e permanent.e en cours d’approbation (pas encore au Canada)', 'Permanent Resident, Pending Approval (Not yet in Canada)', inplace=True)
                df['Status in Canada'].replace('d’approbation (pas encore au Canada)', 'Permanent Resident, Approved in Principle (Not yet in Canada)', inplace=True)
                return df
            
            # Function to extract year from a date string
            def extract_year(date_str):
                # Define regex pattern to extract year
                year_pattern = r'\b(?:19|20)\d{2}\b'  # Matches years in the range 1900-2099

                # Find all matches of year pattern in the date string
                years = re.findall(year_pattern, str(date_str))

                # If at least one year is found, return the first one
                if years:
                    return int(years[0])
                else:
                    return None  # Return None if no year is found
            
            # Function to clean country of birth
            def clean_country_of_birth(country):
                # Check if the entry contains only letters and spaces
                if re.match(r'^[a-zA-Z\s]+$', str(country)):
                    return country
                else:
                    return 'unspecified'
            
            # Function to clean phone number
            def clean_phone_number(Phone):
                Phone = ''.join(filter(str.isdigit, str(Phone)))
                if Phone.startswith('1') and len(Phone) == 11:
                    Phone = Phone[1:]
                elif Phone.startswith('91') and len(Phone) == 12:
                    Phone = Phone[2:]
                if len(Phone) == 10:
                    Phone = f'{Phone[:3]}-{Phone[3:6]}-{Phone[6:]}'
                return Phone
            
            # Function to calculate average age
            def calculate_average_age(age):
                if isinstance(age, float):
                    return None
                if 'Unspecified' in age or '-' not in age:
                    return None
                lower, upper = map(int, age.split('-'))
                return (lower + upper) / 2
            
            # Function to categorize sentiment
            def categorize_sentiment(score):
                if score > 0.2:
                    return 'Positive'
                elif score < -0.2:
                    return 'Negative'
                else:
                    return 'Neutral'

            # Function to perform sentiment analysis on 'UserAnswer' column
            def analyze_sentiment(text):
                # Convert numbers to sentiment categories and set custom sentiment scores
                if isinstance(text,str):
                    if text.isdigit():
                        score = int(text)
                        if score <= 2:
                            return -0.5  # Negative sentiment
                        elif score == 3:
                            return 0.0  # Neutral sentiment
                        elif score >= 4:
                            return 0.5  # Positive sentiment
                    else:
                        # Perform sentiment analysis using TextBlob
                        return TextBlob(str(text)).sentiment.polarity
                else:
                    return None
                
            # Adjust misaligned columns if necessary
            df = adjust_columns(df)
            df = french_convert(df)
            
            # Extract date information
            df[['Date', 'Time']] = df['CompletionDate'].str.split(' ', expand=True)
            df[['Year', 'Month', 'Day']] = df['Date'].str.split('-', expand=True)
            df[['Year', 'Month', 'Day']] = df[['Year', 'Month', 'Day']].apply(pd.to_numeric, errors='coerce')
            df['DayOfWeek'] = pd.to_datetime(df['CompletionDate']).dt.day_name()
            df['WeekNumber'] = pd.to_datetime(df['CompletionDate']).dt.isocalendar().week
            df['Month'] = pd.to_datetime(df['CompletionDate']).dt.month_name()
            df['Quarter'] = pd.to_datetime(df['CompletionDate']).dt.to_period('Q')
            
            # Extract cohort information
            df[['cohort', 'cohort_month_range', 'cohort_year']] = df['Cohort Name'].str.extract(r'([\w\s-]+)\s(\w{3}\s?-?\s?\w{3}),?\s(\d{4})?', expand=True)
            df['cohort_month_range'] = df['cohort_month_range'].str.replace('\s+', ' ')
            date_parts = df['cohort_month_range'].str.extract(r'(\w{3})\s?-?\s?(\w{3})')
            df['cohort_month_start'] = date_parts[0]
            df['cohort_month_end'] = date_parts[1]
            
            # Remove numbering from specified columns
            columns_to_process = ['What is your current employment status', 'What industry or sector are you working in', 'Is this your preferred industry or sector', 'What is your current annual employment salary']
            for column in columns_to_process:
                df[column] = df[column].str.replace(r'^\d+\) ', '', regex=True)
    
            # Applying functions
            df['average_age'] = df['Age'].apply(calculate_average_age)
            df['standardized_phone'] = df['Phone number'].apply(clean_phone_number)
            df['UserAnswerSentiment'] = df['UserAnswer'].apply(analyze_sentiment)
            df['SentimentCategory'] = df['UserAnswerSentiment'].apply(categorize_sentiment)
            
            # Perform sentiment analysis on each row of the 'UserAnswer' column and extract sentiment tags
            def extract_sentiment_tags(text):
                if isinstance(text, str):
                    tags = []
                    sentences = text.split('.')  # Split text into sentences
                    for sentence in sentences:
                        if sentence.strip():  # Check if the sentence is not empty
                            words = [word for word in sentence.split() if TextBlob(word).sentiment.polarity != 0]
                            if words:  # Check if words with non-zero polarity exist in the sentence
                                tags.extend(words)
                    return ','.join(tags)
                else:
                    return None

            # Add a new column for sentiment tags
            df['Tags'] = df['UserAnswer'].apply(extract_sentiment_tags)

            
            # Fill NaN values with 'unspecified' and append
            df.fillna("unspecified", inplace=True)
            # Create a boolean mask to identify rows where "cohort" is "unspecified"
            mask = df["cohort"] == "unspecified"

            # Use the boolean mask to assign values from "Cohort Name" to "cohort" where "cohort" is "unspecified"
            df.loc[mask, "cohort"] = df.loc[mask, "Cohort Name"]
            dfs.append(df)
            
        else:
            print(f"File not found: {file_path}")
    
    # Concatenate all DataFrames into a single DataFrame
    processed_data = pd.concat(dfs, ignore_index=True)               
    return processed_data

# Example usage:
base_directory = r'C:\Users\LENOVO\Desktop\NEU\0. Courses\12. Capstone\Data\Survey Report Assessment'

file_names = []
folder_files = os.listdir(base_directory) 

for file in folder_files:
    # Action with these files
    file_names.append(file)

survey_data = cleaned_survey_report_data(base_directory, file_names)

# Save the processed data to a CSV file
#survey_data.to_csv('survey_data.csv', index=False)

# Display the processed DataFrame
survey_data.head(3)
```
